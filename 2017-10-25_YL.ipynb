```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
def plotStartLocation(df, lon_lbound=-74.1, lon_rbound=-73.85,
                      lat_lbound=40.65, lat_rbound=40.85):
    """
    Plot the start locations of the trips
    """
    df = next(df)
    df = df[(df.Start_Lon < lon_rbound) & (df.Start_Lon > lon_lbound)
            & (df.Start_Lat < lat_rbound) & (df.Start_Lat > lat_lbound)]
    plt.plot(df.Start_Lon, df.Start_Lat, '.', alpha=0.1)
    plt.axis('equal')
```

```python
df = pd.read_csv("data/yellow_tripdata_2009-01.csv", chunksize=5000)
plotStartLocation(df)
```
